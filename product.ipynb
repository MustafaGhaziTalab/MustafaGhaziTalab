```python
!pip install Pillow --upgrade  # تحديث Pillow لدعم WebP

import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# تحميل البيانات من ملف CSV
data = pd.read_csv('amazon.csv')

# تحويل عمود 'rating_count' إلى نوع عددي وتصفية الصور الغير صالحة
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
data = data.dropna(subset=['img_link'])

# تحديد الـ10 منتجات الأعلى تقييمًا
top_10_products = data.nlargest(10, 'rating_count')

# عرض الصور
fig, axes = plt.subplots(2, 5, figsize=(20, 10))
for i, (idx, row) in enumerate(top_10_products.iterrows()):
    try:
        response = requests.get(row['img_link'], timeout=5)
        response.raise_for_status()  # سيتم تشغيل استثناء لأي رد غير ناجح
        img = Image.open(BytesIO(response.content))
        ax = axes[i // 5, i % 5]
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(row['product_name'])
    except Exception as e:
        print(f"An error occurred with image {row['img_link']}: {e}")
        axes[i // 5, i % 5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the data from a CSV file
data = pd.read_csv('amazon.csv')

# Convert related columns to numeric types, handling any conversion errors
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce').fillna(0)
data['rating'] = pd.to_numeric(data['rating'], errors='coerce').fillna(0)
data['discount_percentage'] = pd.to_numeric(data['discount_percentage'], errors='coerce').fillna(0)

# Filter the data to get the top 10 best-selling products based on 'rating_count'
top_10_products = data.nlargest(10, 'rating_count')

# Calculate the average rating and discount percentage
average_rating = top_10_products['rating'].mean()
average_discount = top_10_products['discount_percentage'].mean()

print(f"The average rating of the top 10 best-selling products is: {average_rating:.2f}")
print(f"The average discount percentage of the top 10 best-selling products is: {average_discount:.2f}%")
```
